```python
import numpy as np

class DistanceBasedClassifier:
    def __init__(self, labels=2):
        self.num_labels = labels
        self.centers = {}

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)

        unique_labels = np.unique(y)
        print("Classes:", unique_labels)

        # Compute centroid for each class
        for label in unique_labels:
            self.centers[label] = np.mean(X[y == label], axis=0)

    def predict(self, X):
        X = np.array(X)
        predictions = []

        for x in X:
            predictions.append(self.calculate_distance(x))

        return np.array(predictions)

    def calculate_distance(self, x):
        distances = {}

        # Compute distance from each class center
        for label, center in self.centers.items():
            distances[label] = np.linalg.norm(x - center)

        # Return label with minimum distance
        return min(distances, key=distances.get)
```
